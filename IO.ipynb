# Reading images

```python
import matplotlib.pyplot as plt
%matplotlib inline
from myshow import myshow

filename = 'images/dojo.jpg'
```

```python
# SimpleITK
import SimpleITK as sitk 

image = sitk.ReadImage(filename, sitk.sitkVectorUInt8)
myshow(image)
```

```python
# image origin
myshow(image[0:100,0:100])  
```

```python
import skimage.io as skio

array = skio.imread(filename)
myshow(array)
```

```python
myshow(array[0:100,0:100,:])
```

```python
# The curse of BGR: http://www.learnopencv.com/why-does-opencv-use-bgr-color-format/

import cv2

array2 = cv2.imread(filename)
myshow(array2)
```

```python
array2 = cv2.cvtColor(cv2.imread(filename), cv2.COLOR_BGR2RGB)
myshow(array2)
```

```python
myshow(array2[0:100,0:100,:])
```

# Writing images

```python
import numpy as np
array = np.random.randint(0, 255, size=(100, 100, 3), dtype=np.uint8)
```

```python
myshow(array)
```

```python
# The curse of multidimentions: SimpleITK
sitk.WriteImage(sitk.GetImageFromArray(array, isVector=True), 'outputs/mysitkimage.png');
skio.imsave('outputs/myskimage.png', array);
cv2.imwrite('outputs/mycv2.png', array);
```
